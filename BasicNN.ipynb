```python
import numpy as np
```

```python
xor_input = [[0,0], [0,1], [1,0], [1,1]]
xor_output = [0,1,1,0]
```

```python
x1 = 0
x2 = 1
#Input Matrix
x_m = [x1, x2]

#Layer 1
#Init random weights & bias
w_m = np.random.rand(2,2)
b_m = np.random.rand(1,2)

#Init random weights & bias
w_output = np.random.rand(2,1)
b_output = np.random.rand(1,1)

learning_rate = 0.1
epochs = 3

for c in range(epochs):
    print(f'EPOCH {c}')
    print(f'WEIGHTS AT CURRENT EPOCH   :\nw_output = {w_output}\nw_m = {w_m}')
    print(f'BIASES AT CURRENT EPOCH   :\nb_output = {b_output}\nb_m = {b_m}\n\n')
    for i in range(len(xor_input)):
        
        x_m = xor_input[i] #Shape (1,2)
        y_true = xor_output[i] #Shape (1,1)
        # x@W + b
        layer1_output = x_m @ w_m + b_m #Output Matrix Shape: (1, 2)
        layer1_output_relu = relu(layer1_output) #Output Matrix Shape: (1, 2)

        # Getting the output
        # x@W + b
        final_output = layer1_output_relu @ w_output + b_output #Shape (1,1)
        y_pred = sigmoid(final_output) #Shape (1,1)

        # BACK PROPOGATION

        # BCE loss = -(y_true * log(y_pred) + (1 - y_true) * log(1 - y_pred))
        # BCE loss derivative:
        dL_dy_pred = (y_pred - y_true) / (y_pred * (1 - y_pred))

        # sigmoid derivative
        dy_pred_dfinal_output = y_pred * (1 - y_pred)
        dfinal_output_dw_output = layer1_output_relu #b_output = 0 cos constant

        # Final Gradient for Weights at output
        dL_dw_output =  dfinal_output_dw_output.T * (dL_dy_pred * dy_pred_dfinal_output) # check report for explaination

        # Getting gradient for bias at output
        dfinal_output_db_output = final_output #Shape (1,1)

        # Final Gradient for Bias at output
        dl_db_output = dfinal_output_db_output * (dL_dy_pred * dy_pred_dfinal_output)
        print(f'dl_db_output: {dl_db_output}')
        
        # Hidden Layer
        dfinal_output_dlayer1_output_relu = w_output  # Shape = (2,1)
        dlayer1_output_relu_dlayer1_output = relu_derivative(layer1_output)  # Shape = (1,2)

        dlayer1_output_dw_m = np.array(x_m).reshape(1, 2).T  # Shape = (1,2)

        # Final Gradient for Weights at Hidden Layer 
        # shape =  1 * 1 * ((2,1)*(1,2)) * (1,2) 
        dl_dw_m = (dL_dy_pred *  dy_pred_dfinal_output)  * (dfinal_output_dlayer1_output_relu @ dlayer1_output_relu_dlayer1_output ) @ dlayer1_output_dw_m

        # Getting gradient for bias at Hidden Layer 
        dlayer1_output_db_m = layer1_output.T #Output Matrix Shape: (1, 2)

        # Final Gradient for Bias at output 
        # Shapes = 1 * 1 * (2,1) * (1,2)
        dl_db_m = (dL_dy_pred *  dy_pred_dfinal_output)  * ((dfinal_output_dlayer1_output_relu @ dlayer1_output_relu_dlayer1_output ) @ dlayer1_output_db_m)
        print(f'dl_db_m {dl_db_m}')
        # GRADIENT DESCENT:
        w_output = w_output - dL_dw_output
        w_m -= dl_dw_m

        b_output = b_output - dl_db_output
        b_m = b_m - dl_db_m


        # print(f'\nw_output = {w_output}\nw_m = {w_m}\n')
    
```

```python
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

def sigmoid(x):
    return 1 / (1 + np.e**(-x))
```
